```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report,recall_score,precision_score,f1_score
```

```python
data = pd.read_csv('car_evaluation.csv')
```

```python
data.shape
```

```python
data.head()
```

```python
data.outcome.value_counts()
```

```python
x = data.iloc[0:, 0:6]
x.head()
```

```python
y = data.iloc[0:,-1]
y.head()
```

```python
label = LabelEncoder()
l1 = label.fit_transform(x['lug_boot'])
x['lug_boot'] = l1

l2 = label.fit_transform(x['safety'])
x['safety'] = l2

l3 = label.fit_transform(x['maint'])
x['maint'] = l3

l4 = label.fit_transform(x['buying'])
x['buying'] = l4
```

```python
x.head()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=10)
```

```python
model = KNeighborsClassifier()
model.fit(x_train,y_train)
y_predict = model.predict(x_test)
```

```python
print(confusion_matrix(y_test,y_predict))
print("Accuracy: ",accuracy_score(y_test,y_predict))
print("Precision Score: ", precision_score(y_test,y_predict,average='macro'))
print("Recall score: ", recall_score(y_test,y_predict,average='macro'))
print(classification_report(y_test,y_predict))
```

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE()
```

```python
x_train_smote, y_train_smote = smote.fit_resample(x_train.astype('float'), y_train)
```

```python
from  collections import Counter
print("Before SMOTE :",Counter(y_train))
print("After SMOTE :",Counter(y_train_smote))
```

```python
model.fit(x_train_smote,y_train_smote)
print(confusion_matrix(y_test,y_predict))
print("Accuracy: ",accuracy_score(y_test,y_predict))
print("Precision Score: ", precision_score(y_test,y_predict,average='macro'))
print("Recall score: ", recall_score(y_test,y_predict,average='macro'))
print(classification_report(y_test,y_predict))
```
