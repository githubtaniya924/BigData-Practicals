```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import SelectKBest
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
data = pd.read_csv("mobile_train.csv")
```

```python
data.head()
```

```python
data.shape
```

```python
x = data.iloc[0:, 0:20]
x.head()
```

```python
y =  data.iloc[0:, -1]
y.head()
```

```python
# Standardization is done to give equal importance to the data
# Initialize the scaler
scaler = StandardScaler()

# Fit and transform the data
x_scaled = scaler.fit_transform(x)

#print("Standardized Data (Z-score Normalization):")
#print(x_standardized) -> gives list of numpy array

x = pd.DataFrame(x_scaled, columns=x.columns)
x.head()
```

```python
best_features = SelectKBest()
best_features
```

```python

best_features = SelectKBest(k=10)
fit = best_features.fit(x, y)
fit.scores_
```

```python
dfscores = pd.DataFrame(fit.scores_)
dfcolumns = pd.DataFrame(x.columns)
```

```python
featureScores = pd.concat([dfcolumns, dfscores], axis=1)
featureScores.columns = ['Features', 'Scores']
featureScores = featureScores.sort_values(by='Scores', ascending=False)
featureScores
```

```python
top10_scores = list(featureScores[:10]['Features'].values)
top10_scores
```

## Accuracy Comparison

```python
model = LogisticRegression()
model.fit(x, y)
```

```python
from numpy import average
from sklearn.model_selection import cross_val_score
scores = cross_val_score(model, x, y, cv=5)
print("Cross-Validation Scores:", scores)
average = scores.mean()
print("Average Cross-Validation Score:", average)
```

```python
top_10_scores = cross_val_score(model, x[top10_scores], y, cv=5)
print("Cross-Validation Scores:", top_10_scores)
top10_avg = top_10_scores.mean()
print("Average Cross-Validation Score:", top10_avg)
```

**Conclusion: If the model is given only required no. of features then the accuracy increases
But if unwanted features are also provided then accuracy decreases.**
